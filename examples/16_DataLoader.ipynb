```python
import torch as t
import torch.utils.data as data
import numpy as np

from torch.utils.data import TensorDataset
```

```python
# randn(size)
data_ = t.arange(0,600).view(12,2,5,5)
data_
```

```python
# randint(시작,끝,(size))
labels = t.randint(1,100,(12,))
labels
```

```python
# Must match the size[0] between data & labels
dataset = TensorDataset(data_, labels)
# next(iter(dataset))은 data set을 한개씩 보여주는 역할을 한다.
# 600개의 데이터에서 labels를 12개로 나눴기 때문에, 50개의 데이터가 총 12번씩 보이게 된다.
# 아래는 50, 50, 50 ... 개의 세트중 한개의 data set을 보여주는 것이고
# 맨 아래의 tensor는 label을 뜻한다.
# 최종적으로는 (50,1) = (data set, 레이블) x 12 개의 세트가 나온다.
next(iter(dataset))
```

```python
# Size = (12,2,5,5)
# Size[0]/batch_size 갯수만큼 data set가 나오게 된다.
dataloader = t.utils.data.DataLoader(dataset, batch_size=12, shuffle=True)
```

```python
for i, (batch,labelss) in enumerate(dataloader):
    print(f"{i}".center(100,"="))
    print("Batch size : ", batch.size())
    print("Batch : ", batch)

    print("labels size : ", labelss.size())
    print("Labels : ", labelss)
    
print(f"Finish".center(100,"="))
print("Total Batch Sets : ", i + 1)
```
