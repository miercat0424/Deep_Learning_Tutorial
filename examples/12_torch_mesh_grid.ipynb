```python
import matplotlib.pyplot as plt
import torch as t 
xs = t.linspace(-5,5, steps=100)
ys = t.linspace(-5,5, steps=100)

print(' xs & ys '.center(50,"="))
print(xs)
```

```python

x,y = t.meshgrid(xs,ys, indexing='xy')
z = t.zeros_like(x) + 2

print(' x '.center(100,"="))
print(x)
print(' y '.center(100,"="))
print(y)
print(' z '.center(100,"="))
print(z)
print(' shapes '.center(100,"="))
print(f"x : {x.shape}, y : {y.shape}, z : {z.shape}")
print("="*100)

ax = plt.axes(projection='3d')
ax.plot_surface(x.numpy(), y.numpy(), z.numpy())
ax.scatter(0,0,2,c='r')
plt.show()
```
